The following commands generate random points on a segment of length L, give them a fixed velocity, and estimate the escape probability as a function of time

```python
# The following statements are meant to exploit functionalities to deal with numerical methods and to visualize results
import numpy as np
import matplotlib.pyplot as plt
```

```python
L=10.
v0=2.
```

```python
x=np.random.uniform(0.,L) # This command generates a single random number in the interval [0,L]
print(x)
```

In the following we will generate nsamples different points on the segment and we will compute the probability of escaping from the avearge on these points

```python
nsamples=100000
x=np.random.uniform(0.,L,nsamples)
v=np.random.choice((-1,1),nsamples)*v0 
# choice randomly selects between the provided alternatives, it is used here to generate both positive and negative velocities
```

```python
escape_times_right=((L-x)/v)*(v>0)
escape_times_left=(-x/v)*(v<0)
escape_times = escape_times_left + escape_times_right
```

We can visualize the distribution of escape times using the hist() function of matplotlib

```python
plt.hist(escape_times,bins=100)
plt.show()
```

```python
escape_times.mean()
```
